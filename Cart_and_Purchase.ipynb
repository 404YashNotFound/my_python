# Online Shopping Cart System

Simulate an online shopping cart where a user has a budget of $1000.

    They can add or remove items from the cart.
    Display their remaining balance after each action.
    Allow only 5 purchases before checkout.

```python
# Set some details for the user
budget = 1000  # Initial budget amount
apple = 500    # Price of an apple
banana = 60    # Price of a banana
orange = 250   # Price of an orange
cart_amount = []  # List to store purchased items

print("Welcome to the Online store")
print(f"Your starting budget: ${budget}\n")

for i in range(1, 6):  # Loop to allow up to 5 transactions
    print("1. Add an item")
    print("2. Remove an item")
    print("3. Check balance")
    print("4. Exit")

    choice = input("Enter your choice (1/2/3/4): \n")

    if choice == "1":  # To add an item to the cart
        print("1. Apple")
        print("2. Banana")
        print("3. Orange")
        print("4. Exit")

        User_Input = input("\nWhat product do you want in your cart (1/2/3/4): ")

        # Check the selected item and add to cart if within budget
        if User_Input == "1":  # Apple
            if apple > budget:
                print("Insufficient funds")
            else:
                cart_amount.append(apple)
                budget -= apple
                b = 5 - i
                print(f"Item added! Remaining balance: ${budget}")
                print(f"You have {b} purchases left.\n")

        elif User_Input == "2":  # Banana
            if banana > budget:
                print("Insufficient funds")
            else:
                cart_amount.append(banana)
                budget -= banana
                b = 5 - i
                print(f"Item added! Remaining balance: ${budget}")
                print(f"You have {b} purchases left.\n")

        elif User_Input == "3":  # Orange
            if orange > budget:
                print("Insufficient funds")
            else:
                cart_amount.append(orange)
                budget -= orange
                b = 5 - i
                print(f"Item added! Remaining balance: ${budget}")
                print(f"You have {b} purchases left.\n")

        elif User_Input == "4":  # Exit item selection
            print("Exiting item selection.\n")
        else:
            print("Invalid choice.\n")

    elif choice == "2":  # To remove an item from the cart
        print("1. Apple")
        print("2. Banana")
        print("3. Orange")
        print("4. Exit")

        User_Input2 = input("\nWhat product do you want to remove from your cart (1/2/3/4): ")

        if not cart_amount:  # Check if the cart is empty
            print("Item not in cart")
        else:
            # Check which item to remove and update budget
            if User_Input2 == "1" and apple in cart_amount:
                cart_amount.remove(apple)
                budget += apple
            elif User_Input2 == "2" and banana in cart_amount:
                cart_amount.remove(banana)
                budget += banana
            elif User_Input2 == "3" and orange in cart_amount:
                cart_amount.remove(orange)
                budget += orange
            else:
                print("Item not found in cart.")
                continue  # Skip to the next iteration

            b = 5 - i
            print(f"Item removed! Remaining balance: ${budget}")
            print(f"You have {b} purchases left.\n")

    elif choice == "3":  # To check remaining balance
        print(f"Your remaining balance: ${budget}\n")

    elif choice == "4":  # To exit the program
        print("Thank you for shopping with us!\n")
        break

    else:
        print("Invalid choice. Please select 1, 2, 3, or 4.\n")

print("Final cart items:", cart_amount)  # Display final cart items4
```
